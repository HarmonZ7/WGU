Below I begin by importing pandas and creating a base data frame, "df", and populating it with the provided data set.

```python
import pandas as pd
df = pd.read_excel('D598_Data_Set.xlsx')
df.head(5)
```

Next, I clean the columns and ensure there are no duplicates. In this step I reformat all columns to lowercase and separated by underscores. The output of the block below shows that the columns names were adjusted, and no duplicated business ID's were found.

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
df = df.rename(columns={"total_long-term_debt" : "total_long_term_debt"})
df[df.duplicated("business_id")]
```

The next step was to create a data frame that is grouped by business state and provide descriptive statistics. These statistics include mean, median, minimum value, and maximum value. These are all included in the ".describe()" method used below.

```python
df_statistics = df.groupby("business_state").mean(numeric_only=True)
df_statistics.describe()
```

The next step involved creating a new data frame that highlights all businesses with a negative debt to equity ratio. This is accomplished by populating this new data frame, "df_neg_debt_equity", with all entries in our main df where the column "debt_to_equity" is below zero. The output of this code was filtered to only include relevant columns to this aspect of the project.

```python
df_neg_debt_equity = df[df["debt_to_equity"] < 0]
df_neg_debt_equity[["business_id", "business_state", "debt_to_equity"]]
```

One of the last steps in this project was to create a last data frame that uses data from our original df to calculate each businesses debt to income ratio. I accomplished this by initializing a new data frame and populating that with a dictionary that creates a new column, "debt_to_income", and filling that column with the calculated values from dividing our long term debt by our revenue. I then added the business ID as the foreign key. The output displays the calculated debt to income ratio for each respective business ID.

```python
df_debt_income = pd.DataFrame({"debt_to_income": df["total_long_term_debt"] / df["total_revenue"]})
df_debt_income["business_id"] = df["business_id"]
df_debt_income.head(5)
```

The last step involved adding the calculated "debt_to_income" column to our main df. I accomplished this by utilizing the .concat() method and adding the "debt_to_income" column from our df from the previous step. Lastly, for clarity, I organize the df by state and display in the output how the column was indeed added to our newly combined "df_combined".

```python
df_combined = pd.concat([df, df_debt_income["debt_to_income"]], axis=1)
df_combined.sort_values("business_state").head(5)
```
